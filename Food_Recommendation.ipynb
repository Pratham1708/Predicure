```python
def calculate_bmr(weight, height, age, gender):
    if gender == 'male':
        bmr = 10 * weight + 6.25 * height - 5 * age + 5
    elif gender == 'female':
        bmr = 10 * weight + 6.25 * height - 5 * age - 161
    else:
        print('Invalid gender')
        return None
    return bmr


def calculate_calories(bmr, activity_level):
    if activity_level == 'sedentary':
        pal = 1.2
    elif activity_level == 'lightly active':
        pal = 1.375
    elif activity_level == 'moderately active':
        pal = 1.55
    elif activity_level == 'very active':
        pal = 1.725
    elif activity_level == 'extra active':
        pal = 1.9
    else:
        print('Invalid activity level')
        return None
    calories = bmr * pal
    return calories


weight = float(input('Enter your weight in kilograms: '))
height = float(input('Enter your height in centimeters: '))
age = int(input('Enter your age: '))
gender = input('Enter your gender (male/female): ')

bmr = calculate_bmr(weight, height, age, gender)
if bmr is not None:
    print(f'Your basal metabolic rate is: {bmr:.2f} calories')

print('Select your activity level:')
print('1. Sedentary (little or no exercise)')
print('2. Lightly active (light exercise/sports 1-3 days/week)')
print('3. Moderately active (moderate exercise/sports 3-5 days/week)')
print('4. Very active (hard exercise/sports 6-7 days/week)')
print('5. Extra active (very hard exercise/sports or physical job)')
level = int(input())

if level == 1:
    activity_level = 'sedentary'
elif level == 2:
    activity_level = 'lightly active'
elif level == 3:
    activity_level = 'moderately active'
elif level == 4:
    activity_level = 'very active'
elif level == 5:
    activity_level = 'extra active'
else:
    print('Invalid activity level')
    activity_level = None

if activity_level is not None:
    calories = calculate_calories(bmr, activity_level)
    if calories is not None:
        print(f'Your daily calorie requirement is: {calories:.2f} calories')
```

```python
diseases_data = pd.read_csv(r'/content/drive/MyDrive/Practicum/diseases.csv')

pre_meal = input("Do you prefer Vegetarian or Non-vegetarian meal?: ")
num_diseases = int(input('How many diseases are you suffering from?'))

user_diseases = []
for i in range(num_diseases):
    disease = input("Enter the name of disease:" + str(i+1) + ": ")
    user_diseases.append(disease)

nutritional_components = []
for disease in user_diseases:
    row = diseases_data.loc[diseases_data['Disease'] == disease]
    nutritional_components.append(list(row.iloc[:, 1:].values[0]))

final_list = nutritional_components[0]
for components in nutritional_components[1:]:
    for i, value in enumerate(components):
        final_list[i] = min(final_list[i], value)

food_items_dict = {}
for i, component_value in enumerate(final_list):
    component_name = diseases_data.columns[i+1]
    food_items = []
    for index, row in food.iterrows():
        food_component_value = row[component_name]
        if abs(float(food_component_value) - float(component_value)) / float(component_value) <= 0.90:
            food_items.append(row['food items'])

    food_items_dict[component_name] = food_items

food_i_list = list(food_items_dict.values())
food_i_list = sum(food_i_list, [])
unique_list = list(set(food_i_list))
print("Final List:", final_list)
print("Unique Food Items:", unique_list)
```

```python
import random
from tabulate import tabulate

# Define days of the week
days_of_week = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

# Define meal categories for vegetarian and non-vegetarian meals
def get_meal_categories(pre_meal):
    # Initialize meal_categories to handle cases where pre_meal is invalid.
    meal_categories = {}  
    if pre_meal.lower() == 'vegetarian': # Case-insensitive comparison
        meal_categories = {
            'Breakfast': ['Breakfast grains', 'Fruits', 'Vegetables', 'Protein', 'Healthy Fats', 'Breads', 'Juice', 'Indian bread', 'Tea & Coffee'],
            'Lunch': ['Grains', 'Indian bread', 'Vegetables', 'Salads', 'Healthy Fats', 'Soup', 'Dairy'],
            'Snacks': ['Tea & Coffee', 'Sandwich', 'Nuts & Seeds', 'Fruits', 'Beverages', 'Juice'],
            'Dinner': ['Grains', 'Indian bread', 'Vegetables', 'Salads', 'Healthy Fats', 'Soup', 'Dairy']
        }
    elif pre_meal.lower() == 'non-vegetarian': # Case-insensitive comparison
        meal_categories = {
            'Breakfast': ['Breakfast grains', 'Fruits', 'Vegetables', 'Non-veg Protein', 'Protein', 'Healthy Fats', 'Breads', 'Juice', 'Indian bread', 'Tea & Coffee'],
            'Lunch': ['Grains', 'Indian bread', 'Vegetables', 'Salads', 'Healthy Fats', 'Soup', 'Dairy', 'Meat', 'Non-veg Salads', 'Non-veg Soup'],
            'Snacks': ['Tea & Coffee', 'Sandwich', 'Nuts & Seeds', 'Fruits', 'Beverages', 'Juice', 'Non-veg Sandwich'],
            'Dinner': ['Grains', 'Indian bread', 'Vegetables', 'Salads', 'Healthy Fats', 'Soup', 'Dairy', 'Meat', 'Non-veg Salads', 'Non-veg Soup']
        }
    # If pre_meal is neither "vegetarian" nor "non-vegetarian" (case-insensitive), handle it.
    else:
        print(f"Invalid meal preference: {pre_meal}. Please enter 'Vegetarian' or 'Non-vegetarian'.")
    return meal_categories  

# Define a function to randomly select one dish from each category
def get_random_dish(meal_category, pre_meal):
    meal_categories = get_meal_categories(pre_meal) # Access meal_categories here
    if meal_category in meal_categories:
        return random.choice(meal_categories[meal_category])
    return 'No Dish'

# Function to generate a diet chart for one week
def generate_weekly_diet_chart(pre_meal):
    table = []
    for day in days_of_week:
        row = [day]
        # Randomly select a dish for each meal (Breakfast, Lunch, Snacks, Dinner)
        for meal in ['Breakfast', 'Lunch', 'Snacks', 'Dinner']:
            random_dish = get_random_dish(meal, pre_meal) # Pass pre_meal to get_random_dish
            row.append(random_dish)
        table.append(row)
    return table

# Number of diet charts you want to create
number_of_weeks = 3  # You can change this to generate more tables

# Loop to generate multiple diet charts
for week in range(1, number_of_weeks + 1):
    print(f"\nRandomized Weekly Diet Chart - Week {week}:")
    weekly_diet_chart = generate_weekly_diet_chart(pre_meal) # Pass pre_meal to generate_weekly_diet_chart
    print(tabulate(weekly_diet_chart, headers=['Day', 'Breakfast', 'Lunch', 'Snacks', 'Dinner'], tablefmt='orgtbl'))
    print("\n" + "="*50 + "\n")
```
